```python
#FINDING ANSWERS WITH THE DATA VISUALIZATIONS 

#Q1. What was the best of year of sales?

dataset.groupby("year")['amount'].count().plot(kind="bar",title="Year wise sales")

dataset.groupby("year")['amount'].sum().plot(kind="bar",title="Year wise sales")

dataset.groupby("year")['category'].count().plot(kind="bar",title="Year wise sales")

plt.xlabel("Year")
plt.ylabel("No. of sales")
plt.show()
```

```python
#Q2. BEST MONTH FOR SALES BETWEEN 2015 TO 2018

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset=pd.read_csv('/Users/jasmine/OneDrive/Documents/cleaned.csv')
dataset2015_2018=dataset[(dataset["year"]>= 2015) & (dataset["year"]<= 2018)]

dataset2015_2018.groupby('month')['rating'].count().plot(kind="bar",color="skyblue", edgecolor= "black")

plt.xlabel("Month",fontsize=12)
plt.ylabel("No. of Sales",fontsize=12)
plt.title("Best Month for Sales between 2015-2018",fontsize=18)
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle= "--",alpha= 0.5)
plt.show()
```

```python
#Q3. Highest Selling Year Brand between 2015-2018

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset=pd.read_csv('/Users/jasmine/OneDrive/Documents/cleaned.csv')
dataset2015_2018=dataset[(dataset["year"]>= 2015) & (dataset["year"]<= 2018)]

dataset2015_2018.groupby('brand')['amount'].sum().plot(kind="bar",color="red", edgecolor= "black",label="data")
plt.legend("data")

plt.xlabel("Brand",fontsize=12)
plt.ylabel("Sum of Sales",fontsize=12)
plt.title("Highest Selling Year Brand between 2015-2018",fontsize=18)

plt.grid(axis="y", linestyle= "--",alpha= 0.5)
plt.show()
```

```python
#Q4. BRANDWISE TOP 10 SALES BETWEEN 2015-2018

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset=pd.read_csv('/Users/jasmine/OneDrive/Documents/cleaned.csv')
dataset2015_2018=dataset[(dataset["year"]>= 2015) & (dataset["year"]<= 2018)]

dataset2015_2018.groupby('brand')['amount'].sum().sort_values(ascending=False).head(10).plot(kind="bar",color="red", edgecolor= "black",label="data")
plt.legend("data")

plt.xlabel("Brand",fontsize=12)
plt.ylabel("Total Sales Amount Sales",fontsize=12)
plt.title("Brand wise top 10 sales between 2015-2018",fontsize=18)

plt.grid(axis="y", linestyle= "--",alpha= 0.5)
plt.show()
```

```python
#Q5. What product by category sold the most between 2015 to 2018?

dataset2015_2018=dataset[(dataset["year"]>= 2015) & (dataset["year"]<= 2018)]

dataset2015_2018.groupby('category')['amount'].sum().sort_values(ascending= False).head(10).plot(kind="bar",color="yellow", edgecolor= "black",label="data")
plt.legend("data")

plt.xlabel("Category",fontsize=12)
plt.ylabel("Sum of Sales",fontsize=12)
plt.title("Highest Category sale between 2015-2018",fontsize=18)

plt.grid(axis="y", linestyle= "--",alpha= 0.5)
plt.show()
```

```python
#Q6. Ratings Distribution

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset=pd.read_csv('/Users/jasmine/OneDrive/Documents/cleaned.csv')
sns.countplot(x="rating", data=dataset)
plt.show()
```

```python
#Q7. TOP 10 BEST RATED BRANDS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset=pd.read_csv('/Users/jasmine/OneDrive/Documents/cleaned.csv')
dataset2015_2018=dataset[(dataset["year"]>= 2015) & (dataset["year"]<= 2018)]

dataset2015_2018.groupby('brand')['rating'].mean().sort_values(ascending=False).head(10).plot(kind="bar",color="magenta", edgecolor= "black",title="10 mean")

plt.show()
```

```python
#Q8. Top 5 category sales percentage

c=["purple","yellow","grey","red","skyblue"]
dataset2015_2018.groupby('category')['amount'].sum().sort_values(ascending=False).head(5).plot(kind="pie",autopct="%1.1f%%",startangle=90,color=c)
```

```python
#Q9. Brand wise sales percentage

dataset2015_2018.groupby('brand')['amount'].sum().plot(kind="pie",autopct="%1.1f%%",startangle=90)
```

```python
#Q10. Top 5 Brand wise sales percentage

dataset2015_2018.groupby('brand')['amount'].sum().sort_values(ascending=False).head(5).plot(kind="pie",autopct="%1.1f%%",startangle=90)
```
